## select 명령구문 : select *from 테이블명;
- **DML에 속하는 명령**
- 데이터베이스의 데이터를 읽어올 수 있음.
- SELECT 명령은 '질의' or'쿼리'라 불림
- 애스터리스크(*) = 모든열을 의미하는 메타 문자
- 열(컬럼/필드)과 행(레코드)으로 구성되며, 각 칸을 셀이라 칭함.
- 데이터는 '수치형', '문자열형', '날짜시간형...
- 값이 없는 데이터 = NULL

## 검색 조건 지정: SELECT 열1, 열2 FROM 테이블명 WHERE 조건식

- SELECT 구에서 열 지정하기 = SELECT 열이름1, 열이름2 FROM 테이블명;
- SELECT 구에서 행 출력하기 = SELECT 열 FROM 테이블명 WHERE 조건식;  
ex1) select *from sample21 where no=2;  
ex2) select *from sample21 where name = '박준용'; (문자열도 가능)


- NULL 값 찾을때는 is Null : SELECT *FROM SAMPLE21 WHERE birthday is null;  
- 반대로 Null값 제외할때는 **is not null;**

## 조건 조합하기
- WHERE 조건1 AND 조건2 : 둘다 참이어야 True
- WHERE 조건1 OR 조건2 : 둘중 하나가 참이면 Ture
- WHERE NOT 조건
- AND > OR : AND조건이 우선순위


## 패턴매칭에 의한 검색 : SELECT *FROM 테이블명 WHERE 열명 LIKE '패턴'
- % : 임의의 문자열
- _ : 임의의 하나의 문자  


- **'TEXT%' = 전방일치(지정문자열로 시작)**  
- **'%TEXT%' = 중간일치(지정문자열이 중간에 있음)**  
- **'%TEXT' = 후방일치(지정문자열로 끝)**


- 메타문자 %, _ 를 포함하는 문장의 경우, \을 추출하고 싶은 문자 앞에 삽입: (where 열명 like '%(백슬래시)%%')

## 정렬: SELECT 열명 FROM 테이블명 WHERE 조건식 ORDER BY 열명;

- 오름차순(Default): SELECT 열명 FROM 테이블명 (WHERE 조건식) ORDER BY 열명 ASC;
- 내림차순: SELECT 열명 FROM 테이블명 (WHERE 조건식) ORDER BY 열명 **DESC;**


- 정렬을 하면, 순서대로 보는 효과도 있으면서 비슷한 항목끼리 묶어서 확인이 가능
- 정렬 순서: 영문 -> 숫자 -> 한글


- 복수 열로 정렬 지정: SELECT 열명 FROM 테이블명 **ORDER BY 열명1, 열명2;**
- 내림차순: SELECT 열명 FROM 테이블명 **ORDER BY 열명1 DESC, 열명2 DESC;**

## 결과행 제한하기: LIMIT과 OFFSET을 사용하여 페이지 구분이 가능  

1. 행수 제한: SELECT 열명 FROM 테이블명 WHERE 조건식 ORDER BY 열명 **LIMIT 행수;**  
2. 오프셋 지정: SELECT 열명 FROM 테이블명 ORDER BY 열명 **LIMIT 행수 OFFSET 시작위치;**
    : 시작위치는 0부터 시작!

## 수치연산
- +, -, *, /, %, MOD

- **SELECT *, 식 AS 열 별명 FROM 테이블명;**  
- 열명이 ASCII문자 이외인 경우 **더블 쿼드 ("  ")로 지정!**  
- 연산 후 조건넣어 출력: **SELECT *, 식 AS 열 별명 FROM 테이블명 WHERE 조건식;**  
(조건식은 새로 만든 열별명을 이용할 수 없음.)
- 연산 후 결과값 정렬: **SELECT *, 식 AS 열 별명 FROM 테이블명 ORDER BY 식(열별명)DESC;**  

## 함수
- SELECT **amount, ROUND(amount)** FROM sample34
- 반올림: ROUND(amount, 인수)
- **인수 = 1, 0, -1, -2, -3 : 소수점 둘째자리, 소수점 첫째자리, 1단위, 10단위, 100단위 순서**

## 문자열 연산
1. CONCAT(): 문자열 결합
2. SUBSTRING: 문자열의 일부분을 계산하여 반환 'SELECT *, SUBSTRING(열명, 시작위치, 개수)'
3. TRIM: 스페이스 제거
4. CHAR_LENTH: 문자열의 길이 계산, 가변길이를 갖는 VARCHAR의 경우 문자열 길이 계산이 가능.  
5. OCTET_LENGTH: 문자열의 길이를 바이트 단위로 환산
