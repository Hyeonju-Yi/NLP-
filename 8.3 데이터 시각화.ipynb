## matplotlib으로 그래프 그리기


- %matplotlib qt : 그래프 새로운 창에 생성
- %matplotlib inline : 그래프 현재 창에 생성


- plt.plot(x, y) : 데이터 입력
- plt.show() : 불필요한 정보 삭제하고 그래프만 출력
- plt.figure() : 그래프 창 번호 지정 (첫번재는 figure(1)로 디폴트)
- plt.subplot(m, n, p) : 하나의 그래프 창에 여러개의 하위 그래프를 생성. p 는 그래프가 위치할 하위 그래프 번호
- plt.clf() : 해당 figure의 그래프 모두 삭제


- plt.xlim(xmin, xmax) : 특정범위 확대하여 확인
- plt.xlim(ymin, ymax)

```python
import numpy as np

x = np.arange(-5, 5, 0.1)
y1 = x**2 - 2
y2 = 20*np.cos(x)**2

plt.figure(1)
plt.plot(x, y1)

plt.figure(2)
plt.plot(x, y2)

plt.figure(1)
plt.plot(x, y2) # 1번 그래프 창에 그래프 추가

plt.figure(2)
plt.clf()        # 2번 그래프 창의 모든 그래프를 지움.
plt.plot(x, y1)  # 2번 그래프 창에 새로운 그래프를 생성

plt.show()
```

#### subplot(m, n, p) : 하나의 그래프 창에 m x n 개의 하위그래프 영역 생성

```python
import numpy as np

x = np.arange(0, 10, 0.1)
y1 = 0.3*(x-5)**2 + 1
y2 = -1.5*x + 3
y3 = np.sin(x)**2
y4 = 10*np.exp(-x) + 1

plt.subplot(2,2,1)  # p 는 그래프가 위치할 하위 그래프 번호
plt.plot(x, y1)

plt.subplot(2,2,2)
plt.plot(x, y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)

plt.show()
```

```python
# 그래프 위치(p) 변경

plt.subplot(2,2,2)  # p 는 그래프가 위치할 하위 그래프 번호
plt.plot(x, y1)

plt.subplot(2,2,4)
plt.plot(x, y2)

plt.subplot(2,2,1)
plt.plot(x,y3)

plt.subplot(2,2,3)
plt.plot(x,y4)

plt.show()
```

## 산점도

```python
import matplotlib.pyplot as plt

height = [165, 177, 160, 180, 185, 155, 172]
weight = [62, 67, 55, 74, 90, 43, 64]  

plt.scatter(height, weight)
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.title('Height & Weight')
plt.grid()
```

```python
plt.scatter(height, weight, s = 500, c = 'r')  # 마커 크기는 500, 컬러는 red
plt.show()
```

```python
size = 100 * np.arange(1, 8) # 데이터별로 마커 크기 지정
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y']

plt.scatter(height, weight, s=size, c=colors)
plt.show()
```

```python
import numpy as np

city = ['서울', '인천', '대전', '대구', '울산', '부산', '광주']

lat  = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16] 
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85]

pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995]

size = np.array(pop_den) * 0.2
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y']

plt.scatter(lon, lat, s=size, c=colors, alpha=0.5) # alpha : 투명도
plt.xlabel('경도(longitude)')
plt.ylabel('위도(latitude)')
plt.title('지역별 인구 밀도(2017)')

for x, y, name in zip(lon, lat, city):
    plt.text(x, y, name)  # 마커가 위치한 곳에 도시의 이름 표시
    
plt.show()
```

## 막대그래프

- plt.bar(index, 실제데이터, 컬러, 바이름, 넓이) # 세로 그래프
- plt.barh(index, 실제데이터, 컬러, 바이름, 넓이) # 가로 그래프
- tick_label = 바 밑 라벨 지정


```python
member_IDs = ['m_01', 'm_02', 'm_03', 'm_04'] # 회원 ID
before_ex = [27, 35, 40, 33] # 운동 시작 전
after_ex = [30, 38, 42, 37] # 운동 한 달 후
```

```python
import matplotlib.pyplot as plt
import numpy as np

n_data = len(member_IDs)
index = np.arange(n_data)
plt.bar(index, before_ex)
plt.show()
```

```python
colors = ['r','g','b','m']
plt.bar(index, before_ex, color = colors, tick_label = member_IDs, width = 0.6)
plt.show()
```

```python
colors = ['r','g','b','m']
plt.barh(index, before_ex, color = colors, tick_label = member_IDs)
plt.show()
```

```python
barWidth = 0.4
plt.bar(index, before_ex, color = 'c', align = 'edge', width = barWidth, label = 'before')
plt.bar(index + barWidth, after_ex, color = 'm', align = 'edge', width = barWidth, label = 'after')
       # before_ex 좌표에서 barWidth만큼 우츨으로 이동

plt.xticks(index + barWidth, member_IDs) # x축 tick 라벨 붙이기
plt.legend()                             # 범례 
plt.xlabel('회원 ID')
plt.ylabel('윗몸일으키기 횟수')
plt.title('운동 시작 전과 후의 근지구력(복근) 변화 비교')
plt.show()
```

## 파이 그래프

- plt.pie()
- plt.figure(figsize = (n,n)) 그래프 너비와 높이 설정(1:1 비율로)
- autopct = '%.1f%%'  자동으로 퍼센티지 계산하여 '%'로 표시
- counterclock = False 
- explode : 해당 부분만 분리
- shadow=True : 입체감 생성

```python
import matplotlib.pyplot as plt

fruit = ['사과', '바나나', '딸기', '오렌지', '포도']
result = [7, 6, 3, 2, 2]

plt.pie(result)
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(result, labels=fruit, autopct='%.1f%%')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(result, labels=fruit, autopct='%.1f%%', startangle=90, counterclock=False)
plt.show()
```

```python
explode_value = (0.1, 0, 0, 0, 0)

plt.figure(figsize=(5,5))
plt.pie(result, labels = fruit, autopct='%.1f%%', startangle=90, counterclock=False, explode=explode_value, shadow=True)
plt.show()
```
